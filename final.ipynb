## Exercise 1
Given the following summation, we want to show that if $f_j$ for $j = 0, 1, \dots, N-1$ are real numbers then $c_0$ is real and $c_N-k = \overline{c}_k$ 
>$$c_k = \sum_{j=0}^{N-1}f_je^{-i2\pi kj/N}$$


First, let $k=0$ then $e^{-i2\pi 0j/N} = e^0 = 1$ so our summation becomes $$c_0=\sum_{j=0}^{N-1}f_j.$$
Given that $f_j$ for $j = 0, 1, \dots, N-1$ are real then the summation of real numbers will also be real. Therefore, $c_0$ is real. Next to show that $c_{N-k} = \overline{c}_k$, let's first find what $\overline{c}_k$ is. We can express $\exp{(-i2\pi kj/N)}$ as $\cos(2\pi kj/N) - i\sin(2\pi kj/N)$ which means the complex conjugate is $\cos(2\pi kj/N) + i\sin(2\pi kj/N) = \exp{(i2\pi kj/N)}$. This then gives us our summation for $\overline{c}_k$.

>$$\overline{c}_k = \sum_{j=0}^{N-1}f_je^{i2\pi kj/N} $$

Next, let's examine $c_{N-k}$ by replacing $k$ in our summation with $N-k$.

>$$c_{N-k} = \sum_{j=0}^{N-1}f_je^{-i2\pi (\frac{N-k}{N})j}$$
>$$c_{N-k} = \sum_{j=0}^{N-1}f_je^{i2\pi (\frac{k-N}{N})j}$$
>$$c_{N-k} = \sum_{j=0}^{N-1}f_je^{i2\pi (\frac{k}{N}-1)j}$$
>$$c_{N-k} = \sum_{j=0}^{N-1}f_je^{\frac{i2\pi kj}{N} - i2\pi j}$$
>$$c_{N-k} = \sum_{j=0}^{N-1}f_je^{\frac{i2\pi kj}{N}}e^{- i2\pi j}$$
>$$c_{N-k} = \sum_{j=0}^{N-1}f_je^{\frac{i2\pi kj}{N}}*1$$

Notice that since $j$ is an integer, $e^{-i2\pi j} = \cos(2\pi j) -i\sin(2\pi j) =\cos(2\pi j) -i0 = 1 - 0 = 1 $. Thus $c_{N-k} = \overline{c}_k$ because 
$$c_{N-k} = \sum_{j=0}^{N-1}f_je^{\frac{i2\pi kj}{N}} = \overline{c}_k.$$

```python
#exercise 4
import numpy as np
import math
import matplotlib.pyplot as plt

#keep some global variables for easy access
N = 8
x = [] 
f = []    #f values
f_p = []  #f prime
a = []    #a_k list
b = []    #b_k list
P = []    #P_8(x)
P_p = []  #P_8'(x)

#find x values
print("------------------------x values:------------------------")
for j in range(0, N):
    x.append((j*2*math.pi)/N)
    print("x[%d]"%j, x[j])
print('\n')

#find (f(x_0),...,f(x_7))
print("------------------------f values:------------------------")
for i in range(0, N):
    f.append(math.exp(math.sin(x[i])))
    print("f[x_%d]"%i, f[i])
print('\n')

#find the derivative of f[j]
print("----------------------f prime values:--------------------")
for i in range(0, N):
    f_p.append(math.cos(x[i])*math.exp(math.sin(x[i])))
    print("f'[x_%d]"%i, f_p[i])
print('\n')

#Use numpy FFT package for (d[0], ... , d[7])
A = np.fft.fft(f)

#append a_0
print("------------------------a_k values:----------------------")
a.append(np.real(2 * A[0] /N))
print("a[0]", a[0])

#find a_k and b_k using exercise 2 and 3, also note that from exercise 1 we know the conjugate of A_k is A_{N-k}

for k in range(1, (N//2)+1):
    a.append(np.real((A[k] + A[N-k])/N))
    print("a[%d]"%k, a[k])
print('\n')

print("------------------------b_k values:----------------------")
b.append(0.0)
for k in range(1, N//2):
    b.append(np.real(1.j*(A[k] - A[N-k])/N))
    print("b[%d]"%k, b[k])
print('\n')

#find P_8(x) 
def P_N(a,b,x):
    ak = 0
    bk = 0
    first_term = (1/2)*a[0]
    last_term = ((1/2)*a[N//2]*math.cos((N//2)*x))
    for k in range(1,(N//2)):
        ak += a[k]*math.cos(k*x)
        bk += b[k]*math.sin(k*x)
    return (first_term + ak + bk + last_term)

#Populate our P_8 list 
for num in x:
    P.append(P_N(a,b,num))

print("Here are the values of P_8(x) at x_j for j = 0, 1,..., N-1")
print("-------------------------P values:------------------------")
for i in range(0, len(P)):
    print("P_8[x_%d]"%i, P[i])
print('\n')

#find P_8'(x)
def P_Prime(a,b,x):
    ak = 0
    bk = 0
    last_term = -2*((a[N//2]*np.sin((N//2)*x)))
    for k in range(1,(N//2)):
        ak +=  -k*a[k]*math.sin(k*x) 
        bk += k*b[k]*math.cos(k*x)
    return (ak + bk + last_term)

#Populate our P_8' list
for num in x:
    P_p.append(P_Prime(a,b,num))
    
print("Here are the values of P_8'(x) at x_j for j = 0, 1,..., N-1")
print("-----------------------P Prime values:---------------------")
for i in range(0, len(P_p)):
    print("P_8'[x_%d]"%i, P_p[i])
print('\n')

print("Let E'(x) be the error between f'(x) and P_8'(x): E'(x) = f'(x) - P_8'(x)")
print("------------------------Error values:---------------------")
E_prime = []
for i in range(0, N):
    E_prime.append(f_p[i] - P_p[i])
    print("E'[x_%d]"%i, E_prime[i])
print('\n')

# plt.title("Error graph")
# plt.plot(x, E_prime)        
# plt.show()  
```

By the properties of $\langle\cdot,\cdot\rangle$ we have the following 
>$$\langle f - P_n , q \rangle = \langle f, q \rangle - \langle P_n, q \rangle$$
>$$ = \langle f , \sum_{i=0}^{m}c_i\psi_i(x)\rangle - \langle \sum_{j=0}^{n}a_j\psi_j(x), \sum_{i=0}^{m}c_i\psi_i(x)\rangle $$
>$$ = \sum_{i=0}^{m}(c_i\langle f, \psi_i(x) \rangle) - \sum_{j=0}^{n} \sum_{i=0}^{m}(a_j c_i \langle \psi_j(x), \psi_i(x)\rangle)$$
